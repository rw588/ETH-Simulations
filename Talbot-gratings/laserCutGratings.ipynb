Calculating Talbot length and grating pitch required for interferometry test with light and the 3G interferometer

```python
import numpy as np
import matplotlib.pyplot as plt

def talbot(d,lamda):
    return 2*d**2/lamda

def period(talbotLength,lam, n):
    return np.sqrt(talbotLength*lam/n)
```

```python
ns = np.arange(0.5, 5, 0.5)
wavelengths = np.array([450,632.8])*10**-9

#arrays for period for different n for wavelength in wavelengths
periods = np.array([[period(2.485*10**-3, lam, n) for n in ns] for lam in wavelengths])

#print the shape of periods
print(periods.shape)

# Cap to 3 significant figures
for i in range(len(wavelengths)):
    plt.plot(ns, np.round(periods[i]*10**6, 3), '+', label=f'{wavelengths[i]*10**9:.1f}nm')
plt.xlabel('n \n (number of Talbot lengths between two adjacent gratings)')
plt.ylabel('Period (um)')
plt.legend()
plt.show()
```
